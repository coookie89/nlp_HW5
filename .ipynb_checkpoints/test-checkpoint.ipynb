```python
import json
import random

chinese=[]
english=[]
with open ("./data/json/translation2019zh_valid.json" , 'r') as test_file:
    for times, data in enumerate(test_file, 1):
        data=json.loads(data)
        english.append(data["english"])
        chinese.append(data["chinese"])
        
eng_string=random.choice(english)
print(eng_string)
print(chinese[english.index(eng_string)])
        
        
        
```
